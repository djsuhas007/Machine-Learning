```python
import numpy as np

x = np.linspace(0, 10, 20)

y = x**3

y2 = x**2
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Cubed')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[1].plot(x, y2)
axes[1].set_title('x VS x Squared')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
plt.show()
```

# Combining both the subplots in a single PLOT

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Cubed')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

```python
axes[1].plot(x, y2)
axes[1].set_title('x VS x Squared')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')


plt.show()
```
