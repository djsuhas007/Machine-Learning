```python
import pandas as pd
A = pd.read_csv("E:/Data Science class/Datasets/ts.csv",names=["year","num_pass"])
```

```python
A.year = pd.to_datetime(A.year,format="%Y-%m")
```

```python
A.index = A.year
```

```python
A = A.drop(labels="year",axis=1)
```

```python
import matplotlib.pyplot as plt
plt.plot(A)
plt.xlabel("year")
plt.ylabel("No of passengers travelled")

# Components of TS --> Trend, Season, Error/Rondomness
```

```python
plt.figure(figsize=(10,30))
from statsmodels.api import tsa
tsa.seasonal_decompose(A).plot()
plt.show()
```
