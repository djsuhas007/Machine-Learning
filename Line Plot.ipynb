```python
import numpy as np

x = np.linspace(0, 10, 20)
```

```python
y = x**3
```

```python
y
```

Plot a graph

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
```

```python
plt.title('x vs x Cubed')
```

```python
plt.plot(x, y, 'k')
```

```python
plt.plot(x, y, 'Dk')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.title('x vs x Cubed')
```

```python
plt.plot(x, y, 'D-k')
```

```python
plt.title('x VS x Cubed', fontsize=27)
plt.plot(x, y, 'Dk')
plt.xlabel('Linearly Spaced Numbers')
plt.ylabel('y Value')
plt.title('x vs x Cubed')
plt.plot(x, y, 'D-k')
```
