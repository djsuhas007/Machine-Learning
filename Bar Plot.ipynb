```python
x = ['Shirts', 'Pants','Shorts','Shoes']
```

```python
y = [2000, 2400, 1600, 3600]
```

```python
import matplotlib.pyplot as plt

plt.bar(x, y)
```

```python
plt.title('Sales Revenue by Item Type')
plt.ylabel('Sales Revenue ($)')
plt.xlabel('Item Type')
plt.bar(x, y)
```

For horizontal Bar Graph :

```python
plt.title('Sales Revenue by Item Type')
plt.ylabel('Sales Revenue ($)')
plt.xlabel('Item Type')
plt.barh(x, y)
```
